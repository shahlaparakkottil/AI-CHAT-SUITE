## AI QUESTION ANSWERING SYSTEM(AI Tutor)

```python
from openai import OpenAI
client = OpenAI(
    base_url="http://localhost:11434/v1",api_key="ollama")
print("AI QUESTION ANSWERING SYSTEM(AI Tutor) (type 'exit' to stop)\n")
while True:
    user = input("Enter your coding problem: ")

    if user.lower() == "exit":
        break
    response = client.chat.completions.create(
    model="llama3:8b",
    messages=[
        {"role": "system","content":
         "You are an AI Question answering System .give me the answer for the questions clearly and briefly."},
        {"role": "user",
         "content": user}])

    print("\nGenerated Code:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*60 + "\n")
```
