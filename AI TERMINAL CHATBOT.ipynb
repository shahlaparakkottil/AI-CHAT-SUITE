## AI TERMINAL CHATBOT

```python
from openai import OpenAI
```

```python
client=OpenAI(base_url="http://localhost:11434/v1",api_key="ollama")
```

```python
while True:
    user=input("enter something")
    if user.lower()=="exit":
        break
    reponse=client.chat.completions.create(model="llama3:8b",messages=[{"role":"user","content":user}])
    print(reponse.choices[0].message.content)
```
