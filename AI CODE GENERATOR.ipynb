## AI CODE GENERATOR

```python
from openai import OpenAI
# connect to local ollama model
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)
print("AI Code Generator (type 'exit' to stop)\n")
while True:

    user = input("Enter your coding problem: ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an expert programming assistant. Generate clean and correct code with explanation."
            },
            {
                "role": "user",
                "content": user
            }
        ]
    )

    print("\nGenerated Code:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*60 + "\n")
```
