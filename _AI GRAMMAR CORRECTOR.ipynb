##  AI GRAMMAR CORRECTOR

```python
from openai import OpenAI
client = OpenAI(
    base_url="http://localhost:11434/v1",api_key="ollama")
print(" AI GRAMMAR CORRECTOR (type 'exit' to stop)\n")
while True:
    user = input("Enter your coding problem: ")

    if user.lower() == "exit":
        break
    response = client.chat.completions.create(
    model="llama3:8b",
    messages=[
        {"role": "system","content":
         "You are an  AI grammer corrector. correct the grammers clraerly and proffesionally."},
        {"role": "user",
         "content": user}])

    print("\nGenerated Code:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*60 + "\n")
```
